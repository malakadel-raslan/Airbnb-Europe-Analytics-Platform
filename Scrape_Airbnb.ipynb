```python
import time
import random
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

cities = ['amsterdam', 'athens', 'barcelona', 'berlin', 'budapest', 
            'lisbon', 'london', 'paris', 'rome', 'vienna']

all_data = []

for city_count, city in enumerate(cities, 1):
    print(f"[{city_count}/{len(cities)}] جاري سحب روابط: {city.title()}")
    
    options = Options()
    options.add_argument("--window-size=1200,800")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36")

    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)
    
    try:
        driver.get(f"https://www.airbnb.com/s/{city}/homes")
        time.sleep(random.uniform(8, 12))
        
        listings = driver.find_elements(By.CSS_SELECTOR, "div[data-testid='card-container']")
        print(f"   → تم العثور على {len(listings)} إعلان")
        
        urls = []
        for listing in listings[:25]:   # أول 25 إعلان
            try:
                url = listing.find_element(By.TAG_NAME, "a").get_attribute("href")
                urls.append(url)
            except:
                continue
        
        driver.quit()   # إغلاق المتصفح بعد سحب الروابط
        
        # ====================== مرحلة استخراج البيانات من كل رابط ======================
        print(f"   → بدء استخراج التفاصيل من {len(urls)} رابط...")
        
        for url in urls:
            try:
                driver = webdriver.Chrome(service=service, options=options)
                driver.get(url)
                time.sleep(random.uniform(4, 7))
                
                # الاسم
                try:
                    name = driver.find_element(By.TAG_NAME, "h1").text.strip()
                except:
                    name = "No Title"
                
                # السعر
                price = "N/A"
                try:
                    body_text = driver.find_element(By.TAG_NAME, "body").text
                    price_match = re.search(r'([€$£]\s*[\d,]+)', body_text)
                    if price_match:
                        price = price_match.group(1)
                except:
                    pass
                
                # التقييم + Reviews + Favorite
                rating = "N/A"
                reviews = "0"
                is_fav = 0
                try:
                    body_text = driver.find_element(By.TAG_NAME, "body").text
                    if "Guest favorite" in body_text:
                        is_fav = 1
                    
                    rating_match = re.search(r'(\d\.\d{1,2})', body_text)
                    if rating_match:
                        rating = rating_match.group(1)
                    
                    reviews_match = re.search(r'(\d+)\s*reviews?', body_text, re.I)
                    if reviews_match:
                        reviews = reviews_match.group(1)
                except:
                    pass
                
                all_data.append({
                    "City": city.title(),
                    "Name": name,
                    "Price": price,
                    "Rating": rating,
                    "Reviews": reviews,
                    "Guest_Favorite": is_fav,
                    "URL": url
                })
                
                print(f"     ✓ {name[:50]}... | {price} | ★{rating}")
                
            except:
                pass
            finally:
                try:
                    driver.quit()
                except:
                    pass
            
            time.sleep(random.uniform(3, 6))
            
    except Exception as e:
        print(f"   خطأ في {city}: {str(e)[:100]}")
        if 'driver' in locals():
            driver.quit()

# حفظ
dff = pd.DataFrame(all_data)
dff.to_csv("airbnbb_listings.csv", index=False, encoding='utf-8-sig')

print(f"\n✅ انتهى السكراب!")
print(f"إجمالي الإعلانات: {len(df)}")
print("تم الحفظ: airbnb_listings.csv")
```
